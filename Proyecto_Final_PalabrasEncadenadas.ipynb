 El juego de palabras encadenadas que consiste en buscar la letra final con la que termina una palabra y buscar otra palabra que comience con esta letra para así poco a poco ir formando una cadena.

 Las partidas constan de 10 turnos por jugador, solo pueden jugar dos a la vez.

Se le dará a cada jugador un máximo de 3 intentos o errores por partida, cuando uno de los jugadores exceda esos intentos terminará la partida, al igual que cuando completen ambos las 10 palabras.

```python
#------------------------------------------------------------------------
dict_jugadores = {}
import string
import random
from operator import itemgetter
#-----------------------------------------------------------------------
#TIENE LA FUNCIONALIDAD DE AGREGAR EL JUGADOR AL REGISTRO, SI YA ESTÁ EN EL NO LO DIRÁ
#TAMBIÉN MENCIONA LOS PUNTOS GENERADOS POR EL JUGADOR 
def registro(): 
    nombre_jugador = input("Ingresa el nombre del jugador: ")
    if nombre_jugador not in dict_jugadores.keys():
        dict_jugadores[nombre_jugador] = {
            'puntuacion_general': 0,
            'partidas': []
        }
    else:
        print("Nombre de jugador ya existe")
#--------------------------------------------------------------------------------------------------------------------
#BUSCA QUE EL JUGADOR ESTÉ REGISTRADO, SI ES ASÍ MENCIONA LOS PUNTOS GENERADOS EN LAS PARTIDAS
#(LAS PARTIDAS TIENEN 10 RONDAS) Y ADEMÁS MENCIONA LA POSICIÓN DE JUGADOR EN EL RANKING
def puntuaciones():
    final = True
    while final:
        nombre_jugador = input("Ingresa el nombre del jugador a ver puntuacion: ")
        if nombre_jugador in dict_jugadores:
            list_puntuaciones = list([x, dict_jugadores[x]['puntuacion_general']] for x in dict_jugadores)
            list_puntuaciones = sorted(list_puntuaciones,key=itemgetter(1),reverse=True)
            print("Hola {} tus puntuaciones son las siguientes".format(nombre_jugador))
            partidas = list([x] for x in dict_jugadores[nombre_jugador]['partidas'])
            contador = 1
            for x in partidas:
                print("Partida {} = {}".format(contador,x))
                contador+= 1
            posicion = list(x[0] for x in list_puntuaciones).index(nombre_jugador)+1
            print("Actualmente te encuentras en la posición {} del Ranking.".format(posicion))
            contador = 1
            for x in list_puntuaciones:
                if contador == posicion:
                    print("{}. {} {} puntos <---".format(contador,x[0],x[1]))
                else:
                    print("{}. {} {} puntos".format(contador,x[0],x[1]))
                contador+= 1
            final = False
        else:
            print("Nombre de jugador no encontrado")
#----------------------------------------------------------------------------------------------
#NOS AYUDA A ASIGNAR LOS PUNTOS SEGUN LA CANTIDAD DE LETRAS QUE TIENE NUESTRA PALABRA
#EL MAXIMO DE PUNTOS SON 7 Y EL MINIMO 1
def conseguir_puntos_palabra(palabra):
    if palabra.__len__() > 7:
        return 6
    elif palabra.__len__() == 7:
        return 5
    elif palabra.__len__() == 6:
        return 4
    elif palabra.__len__() == 5:
        return 3
    elif palabra.__len__() == 4:
        return 2
    else:
        return 1
#-------------------------------------------------------------------------------------------
#NOS AYUDA A VERIFICAR QUE LA PALABRA INICIE CON LA LETRA CORRESPONDIENTE Y LLEVAR
#EL CONTEO DE LOS 3 INTENTOS QUE SE TIENEN 
def pedir_palabra(nombre_jugador,letra_actual, errores_jugador):
    jugador_palabra_actual = input("Respuesta {}: ".format(nombre_jugador)).upper().replace(" ", "")
    if jugador_palabra_actual[0] != letra_actual:
        print("Palabra invalida, debe iniciar con la letra {}".format(letra_actual))
        errores_jugador += 1
        if errores_jugador >= 3:
            return False, False
        return pedir_palabra(nombre_jugador,letra_actual, errores_jugador)
    return jugador_palabra_actual, errores_jugador
#--------------------------------------------------------------------------------
#PROGRAMA PRINCIPAL
#PRIMERO VERIFICA QUE LOS JUGADORES ESTÉN REGISTRADOS PARA PODER INICIAR
#NOS AYUDA ASIGNAR LOS TURNOS DE LOS DOS JUGADORES
#VERIFICA QUE SI LA PALABRA YA SE HABÍA USADO, EN CASO DE QUE SÍ, VALE 0 PUNTOS
#TERMINA EL JUEGO SI SE USAN LAS 3 OPORTUNIDADES 

def jugar():
    nombre_jugador_1 = input("Ingresa el nombre del jugador 1: ")
    nombre_jugador_2 = input("Ingresa el nombre del jugador 2: ")
    if nombre_jugador_1 not in dict_jugadores or nombre_jugador_2 not in dict_jugadores:
        print("Al menos un usuario no esta registrado, porfavor registralos antes de jugar")
    else:
        print("Jugador 1= {}".format(nombre_jugador_1))
        print("Jugador 2= {}".format(nombre_jugador_2))
        turno_actual = 1
        errores_jugador_1 = 0
        errores_jugador_2 = 0
        letra_actual = random.choice(string.ascii_letters).upper()
        palabras_usadas = []
        puntos_jugador_1 = 0
        puntos_jugador_2 = 0
        print("{} tu primera palabra debe de iniciar con la letra {}".format(nombre_jugador_1, letra_actual))
        while turno_actual <= 10:
            jugador_1_palabra_actual, errores_jugador_1  = pedir_palabra(nombre_jugador_1, 
                                                                        letra_actual,errores_jugador_1)
            if jugador_1_palabra_actual == False:
                print("Error, terminando juego")
                break
            elif jugador_1_palabra_actual in palabras_usadas:
                print("Palabra usada, 0 puntos")
            else:
                puntos_jugador_1 += conseguir_puntos_palabra(jugador_1_palabra_actual)
                palabras_usadas.append(jugador_1_palabra_actual)
            letra_actual = jugador_1_palabra_actual[-1]

            jugador_2_palabra_actual, errores_jugador_2 = pedir_palabra(nombre_jugador_2, letra_actual,
                                                                        errores_jugador_2)
            if jugador_2_palabra_actual == False:
                print("Error, terminando juego")
                break
            elif jugador_2_palabra_actual in palabras_usadas:
                print("Palabra usada, 0 puntos")
            else:
                puntos_jugador_2 += conseguir_puntos_palabra(jugador_2_palabra_actual)
                palabras_usadas.append(jugador_2_palabra_actual)
            turno_actual+=1
            letra_actual = jugador_2_palabra_actual[-1]
        print("Fin del juego")
        dict_jugadores[nombre_jugador_1]['puntuacion_general'] += puntos_jugador_1
        dict_jugadores[nombre_jugador_1]['partidas'].append(puntos_jugador_1)
        dict_jugadores[nombre_jugador_2]['puntuacion_general'] += puntos_jugador_2
        dict_jugadores[nombre_jugador_2]['partidas'].append(puntos_jugador_2)
#------------------------------------------------------------------
#EL MENÚ NOS DA 4 OPCIONES QUE ES REGISTRAR, VERIFICAR PUNTOS, JUGAR Y SALIR DEL JUEGO 
#ESTAS OPCIONES NOS DIRIGEN A LA PARTE A LA CUAL QUEREMOS ACCEDER 
def menu():
    opcion = 0
    while opcion != 4:
        print("""
        1. Registrar jugadores 
        2. Verificar puntuaciones 
        3. Jugar 
        4. Salir 
            """)
        try:
            opcion = int(input("Ingresa tu opcion: "))
        except ValueError:
            print("Opcion Invalida")
        if opcion == 1:
            registro()
        elif opcion == 2:
            puntuaciones()
        elif opcion == 3:
            jugar()
        elif opcion == 4:
            print("Adios")
        else:
            print("Opcion Invalida")
menu()
```
